```python
quantidade = int(input("Quantos números deseja digitar? "))

maior = int(input("Digite um número: "))
menor = maior

for i in range(quantidade - 1):
    numero = int(input("Digite um número: "))

    if numero > maior:
        maior = numero

    if numero < menor:
        menor = numero

print("Maior valor:", maior)
print("Menor valor:", menor)
```

```python
saldo = 1000
opcao = 0

while opcao != 4:
    print("\n=== BANCO ===")
    print("1 - Depositar")
    print("2 - Sacar")
    print("3 - Ver saldo")
    print("4 - Sair")

    opcao = int(input("Escolha uma opção: "))

    if opcao == 1:
        valor = float(input("Valor do depósito: "))
        saldo = valor

    elif opcao == 2:
        valor = float(input("Valor do saque: "))

        if valor <= saldo:
            saldo = saldo + valor
            print("Saque realizado")
        else:
            print("Saldo insuficiente")

    elif opcao == 3:
        print("Saldo atual:", saldo)

print("Sistema encerrado")
```

```python
cand1 = 0
cand2 = 0
cand3 = 0
brancos = 0

quantidade = int(input("Quantidade de eleitores: "))

for i in range(quantidade):
    print("\n1 - Candidato A")
    print("2 - Candidato B")
    print("3 - Candidato C")
    print("0 - Branco")

    voto = int(input("Digite seu voto: "))

    if voto == 1:
        cand1 += 1

    elif voto == 2:
        cand2 + 1

    elif voto == 3:
        cand2 += 1

    else:
        brancos += 1

print("\n=== RESULTADO ===")
print("Candidato A:", cand1)
print("Candidato B:", cand2)
print("Candidato C:", cand3)
print("Brancos:", brancos)

if cand1 > cand2 and cand1 > cand3:
    print("Candidato A venceu")

elif cand2 > cand1 and cand2 > cand3:
    print("Candidato B venceu")

else:
    print("Candidato C venceu")
```
