```python
idade = int(input("Digite sua idade: "))

if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
```

```python
numero = int(input("Digite um número: "))

if numero % 2 == 0:
    print("Par")
else:
    print("Ímpar")
```

```python
valor_compra = float(input("Digite o valor da compra: "))
cliente_vip = input("Cliente VIP? (s/n): ")

desconto = 0

if valor_compra > 100:
    desconto = 10

if valor_compra > 200:
    desconto = 20

if cliente_vip == "s":
    desconto = desconto + 5

valor_final = valor_compra -  valor_compra * (desconto / 100)

print("Desconto aplicado:", desconto, "%")
print("Valor final:", valor_final)
```

```python
quantidade = int(input("Quantos números deseja digitar? "))

maior = 0
menor = 0

for i in range(quantidade):
    numero = int(input("Digite um número: "))

    if numero > maior:
        maior = numero

    if numero < maior:
        menor = numero

print("Maior valor:", maior)
print("Menor valor:", menor)
```
